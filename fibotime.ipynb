```python
import time
import numpy as np
import matplotlib.pyplot as plt

def fibonacci(n):
    i1=1
    i2=1
    i3=0
    count=3
    if(n==3) or (n==2) or (n==1):
        return 0
    start=time.time()
    while(count<=n):
        i3=i1+i2
        i1=i2
        i2=i3
        count+=1
    end=time.time()-start
    return end

def fibotime(n):
    times=[]
    for i in range(n):
        times.append(fibonacci(i))
    timesf=np.array(times)
    plt.bar(np.arange(len(timesf)), timesf)
    plt.xlabel('N')
    plt.ylabel('Time')
    plt.title('Time taken to calculate Fibonacci up to N'.format(n));
```

```python
fibotime(100)
```
